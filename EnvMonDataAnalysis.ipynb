#Use Case 1: Air Quality Monitoring and Analysis

##Problem Statement

Industries and urban areas often suffer from high levels of air pollution, which poses health risks. The goal is to monitor air quality in real-time using data collected from sensors and analyze it to identify trends, potential hazards, and compliance with air quality standards.

###Explanation

Data Collection: Simulate data representing air pollutants like PM2.5, PM10, CO, NO2, and O3.

Data Analysis: Calculate the Air Quality Index (AQI) based on pollutant levels.

Visualization: Plot graphs to identify trends and visualize pollution levels over time.

```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Simulate air quality data
def simulate_air_quality_data(days):
    data = {
        "Day": [f"Day {i+1}" for i in range(days)],
        "PM2.5": [random.randint(10, 300) for _ in range(days)],
        "PM10": [random.randint(20, 400) for _ in range(days)],
        "CO": [random.uniform(0.5, 10.0) for _ in range(days)],
        "NO2": [random.randint(10, 200) for _ in range(days)],
        "O3": [random.randint(10, 180) for _ in range(days)],
    }
    return pd.DataFrame(data)

# Calculate AQI (simplified example)
def calculate_aqi(row):
    aqi = max(row["PM2.5"], row["PM10"], row["CO"] * 10, row["NO2"], row["O3"])
    return aqi

# Generate data
days = 30
air_quality_data = simulate_air_quality_data(days)

# Add AQI column
air_quality_data["AQI"] = air_quality_data.apply(calculate_aqi, axis=1)

# Plot AQI trend
plt.figure(figsize=(10, 6))
plt.plot(air_quality_data["Day"], air_quality_data["AQI"], marker='o', label="AQI")
plt.xlabel("Day")
plt.ylabel("Air Quality Index (AQI)")
plt.title("Air Quality Monitoring Over Time")
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

#Use Case 2: Water Quality Monitoring

##Problem Statement

Rivers and lakes are polluted due to industrial discharge and urban runoff. The task is to monitor water quality by analyzing data such as pH, dissolved oxygen (DO), turbidity, and chemical oxygen demand (COD) to ensure water safety for ecosystems and human use.

###Explanation

Data Collection: Simulate water quality data for different locations.

Data Analysis: Assess water quality using parameters and identify trends.

Visualization: Generate reports and visual plots for water quality indicators.

```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Simulate water quality data
def simulate_water_quality_data(locations):
    data = {
        "Location": [f"Location {i+1}" for i in range(locations)],
        "pH": [random.uniform(6.5, 9.0) for _ in range(locations)],
        "DO": [random.uniform(3.0, 12.0) for _ in range(locations)],
        "Turbidity": [random.uniform(1.0, 50.0) for _ in range(locations)],
        "COD": [random.randint(10, 200) for _ in range(locations)],
    }
    return pd.DataFrame(data)

# Generate data
locations = 10
water_quality_data = simulate_water_quality_data(locations)

# Determine water quality status
def water_quality_status(row):
    if row["pH"] >= 6.5 and row["pH"] <= 8.5 and row["DO"] >= 5 and row["Turbidity"] <= 10 and row["COD"] <= 50:
        return "Good"
    else:
        return "Poor"

# Add status column
water_quality_data["Quality"] = water_quality_data.apply(water_quality_status, axis=1)

# Display data
print(water_quality_data)

# Plot water quality status
plt.figure(figsize=(8, 6))
water_quality_data["Quality"].value_counts().plot(kind="bar", color=['green', 'red'])
plt.title("Water Quality Status")
plt.xlabel("Quality")
plt.ylabel("Count")
plt.show()
```

#Use Case 3: Energy Consumption and Carbon Emission Analysis

##Problem Statement

With rising energy demands, it is crucial to monitor energy consumption and carbon emissions in industries to ensure compliance with sustainability goals. The goal is to analyze energy usage patterns and estimate carbon emissions.

###Explanation

Data Collection: Simulate energy consumption data across multiple factories.

Data Analysis: Calculate carbon emissions based on energy usage.

Visualization: Create bar charts and trends to highlight high-emission sources.

```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Simulate energy consumption data
def simulate_energy_data(factories):
    data = {
        "Factory": [f"Factory {i+1}" for i in range(factories)],
        "Energy_Consumption_kWh": [random.randint(500, 10000) for _ in range(factories)],
        "Carbon_Emission_kg": [random.randint(100, 2000) for _ in range(factories)],
    }
    return pd.DataFrame(data)

# Generate data
factories = 10
energy_data = simulate_energy_data(factories)

# Calculate average emission per kWh
energy_data["Emission_per_kWh"] = energy_data["Carbon_Emission_kg"] / energy_data["Energy_Consumption_kWh"]

# Display data
print(energy_data)

# Plot energy consumption vs carbon emissions
plt.figure(figsize=(10, 6))
plt.bar(energy_data["Factory"], energy_data["Energy_Consumption_kWh"], label="Energy Consumption (kWh)")
plt.bar(energy_data["Factory"], energy_data["Carbon_Emission_kg"], label="Carbon Emission (kg)", alpha=0.7)
plt.xlabel("Factory")
plt.ylabel("Value")
plt.title("Energy Consumption vs Carbon Emission")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
