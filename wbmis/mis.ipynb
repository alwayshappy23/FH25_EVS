```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import random

# Simulating random data for analysis
data = {
    "Department": ["Cardiology", "Neurology", "Oncology", "Pediatrics", "Orthopedics"],
    "Patients_Online": [random.randint(50, 200) for _ in range(5)],
    "Patients_Offline": [random.randint(30, 150) for _ in range(5)],
}

# Creating a DataFrame
df = pd.DataFrame(data)

# Display the data
print("Simulated Data:")
print(df)

# Plot 1: Bar chart to compare online vs offline patients
plt.figure(figsize=(10, 6))
x = range(len(df["Department"]))
plt.bar(x, df["Patients_Online"], width=0.4, label="Online Patients", color="blue", align="center")
plt.bar(x, df["Patients_Offline"], width=0.4, label="Offline Patients", color="green", align="edge")
plt.xticks(x, df["Department"])
plt.xlabel("Departments")
plt.ylabel("Number of Patients")
plt.title("Comparison of Online vs Offline Patients")
plt.legend()
plt.show()

# Plot 2: Pie chart to show percentage distribution of online patients by department
plt.figure(figsize=(8, 8))
plt.pie(df["Patients_Online"], labels=df["Department"], autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title("Percentage of Online Patients by Department")
plt.show()
```
