```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random

# Simulate air quality data
def simulate_air_quality_data(days):
    data = {
        "Day": [f"Day {i+1}" for i in range(days)],
        "PM2.5": [random.randint(10, 300) for _ in range(days)],
        "PM10": [random.randint(20, 400) for _ in range(days)],
        "CO": [random.uniform(0.5, 10.0) for _ in range(days)],
        "NO2": [random.randint(10, 200) for _ in range(days)],
        "O3": [random.randint(10, 180) for _ in range(days)],
    }
    return pd.DataFrame(data)

# Calculate AQI (simplified example)
def calculate_aqi(row):
    return max(row["PM2.5"], row["PM10"], row["CO"] * 10, row["NO2"], row["O3"])

# Generate data
days = 30
air_quality_data = simulate_air_quality_data(days)
air_quality_data["AQI"] = air_quality_data.apply(calculate_aqi, axis=1)

# Line plot: AQI trend
plt.figure(figsize=(10, 6))
plt.plot(air_quality_data["Day"], air_quality_data["AQI"], marker='o', label="AQI")
plt.xlabel("Day")
plt.ylabel("Air Quality Index (AQI)")
plt.title("AQI Trend Over Time")
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

# Bar chart: Average pollutant levels
average_pollutants = air_quality_data[["PM2.5", "PM10", "CO", "NO2", "O3"]].mean()
average_pollutants.plot(kind="bar", color="skyblue", figsize=(8, 6))
plt.title("Average Pollutant Levels")
plt.ylabel("Level")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Heatmap: Correlation between pollutants
plt.figure(figsize=(8, 6))
sns.heatmap(air_quality_data[["PM2.5", "PM10", "CO", "NO2", "O3", "AQI"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Between Pollutants and AQI")
plt.tight_layout()
plt.show()

# Histogram: AQI distribution
plt.figure(figsize=(8, 6))
plt.hist(air_quality_data["AQI"], bins=10, color="lightgreen", edgecolor="black")
plt.title("Distribution of AQI")
plt.xlabel("AQI")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# Scatter plot: PM2.5 vs AQI
plt.figure(figsize=(8, 6))
plt.scatter(air_quality_data["PM2.5"], air_quality_data["AQI"], color="purple")
plt.title("PM2.5 vs AQI")
plt.xlabel("PM2.5 Level")
plt.ylabel("AQI")
plt.grid()
plt.tight_layout()
plt.show()

# Line plot: Comparison of multiple pollutants over time
plt.figure(figsize=(10, 6))
plt.plot(air_quality_data["Day"], air_quality_data["PM2.5"], label="PM2.5", marker='o')
plt.plot(air_quality_data["Day"], air_quality_data["PM10"], label="PM10", marker='x')
plt.plot(air_quality_data["Day"], air_quality_data["CO"] * 10, label="CO (scaled x10)", marker='s')
plt.plot(air_quality_data["Day"], air_quality_data["NO2"], label="NO2", marker='^')
plt.plot(air_quality_data["Day"], air_quality_data["O3"], label="O3", marker='d')
plt.xlabel("Day")
plt.ylabel("Pollutant Levels")
plt.title("Pollutant Levels Over Time")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
